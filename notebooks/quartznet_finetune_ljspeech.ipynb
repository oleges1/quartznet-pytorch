```
import os
import numpy as np

checkpoint_files = [item for item in os.listdir('checkpoints') if item.endswith('pth')]
epochs = [int(file.split('_')[1]) for file in checkpoint_files]
checkpoint_file = checkpoint_files[int(np.argmax(epochs))]
```
